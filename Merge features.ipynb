```python
import pandas as pd 
import numpy as np 
import matplotlib.pyplot as plt 
```

```python
df_age = pd.read_csv('buurt_leeftijd_bevolkingsdichtheid.csv', index_col=0)
df_building_year = pd.read_csv('buurt_bouwjaar.csv', index_col=0)
df_dummy = pd.read_csv('dummy_def_neighborhood_V2.csv', index_col=0)
```

```python
df_age
```

```python
df_building_year.rename(columns={'Wijken en buurten' : 'BuurtNaam'}, inplace=True)
df_building_year = df_building_year.replace('.', np.NaN)

df_building_year
```

```python
df_dummy.rename(columns={'neighborhood' : 'BuurtNaam'}, inplace=True)
df_dummy
```

```python
df_all_features = pd.merge(df_age, df_building_year, how='outer', on='BuurtNaam')
df_all_features
```

```python
df_all_features = pd.merge(df_all_features, df_dummy, how='outer', on='BuurtNaam')
df_all_features
```

```python
# Write to csv
df_all_features.to_csv('all_features.csv')
```
