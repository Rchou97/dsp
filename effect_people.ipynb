```python
# Load the Pandas libraries with alias 'pd' 
import pandas as pd 
# Read data from file 'filename.csv' 
# (in the same directory that your python process is based)
# Control delimiters, rows, column names with read_csv (see later) 
df = pd.read_csv('final-amsterdamzuid.csv',delimiter=',') 
# Preview the first 5 lines of the loaded data 
#df_kadaster.columns =['id','perceelid', 'perceeltype', 'huisnr', 'huisnr_bag_letter','huisnr_bag_toevoeging','bag_nummeraanduidingid','bag_adresseerbaarobjectid','lat','lon','rdx','rdy','oppervlakte','gebruiksdoel','reeksid']
df = df.set_index('Unnamed: 0')
df.index.names = ['id']
df.head()
```

```python
df['x'].dtype
```

```python
straal = 15

df = df.copy()

for i, j in df.iterrows():
    rdx_punt = j[17]
    rdy_punt = j[18]
    #print("rdx: {} - rdy: {}".format(rdx_punt,rdy_punt))

    rdx_min = rdx_punt - straal
    rdx_max = rdx_punt + straal
    rdy_min = rdy_punt - straal 
    rdy_max = rdy_punt + straal

    rslt_df = df[(df['x'] >= rdx_min) &
               (df['x'] <= rdx_max) &
              (df['y'] >= rdy_min) &
              (df['y'] <= rdy_max)]

    df.at[i,'Buren'] = len(rslt_df.index)
df
```

```python
df_cbs = pd.read_csv('cbs-pc6-2020-v1.csv',delimiter=';',skiprows=7) 
df_cbs = df_cbs.set_index('Postcode')
df_cbs = df_cbs.replace(-99997, 2)
df_cbs
```

```python
## Kadaster x CBR dataset (PC6)

df = df.copy()

results = list()
for i in df.iterrows():
    pc6 = df.loc[i[0], 'postcode']

    if pc6 in df_cbs.index:
        df.at[i[0],'Huishoudgrootte'] = df_cbs.at[pc6,'Huishoudgrootte']
        df.at[i[0],'Bijstand'] = df_cbs.at[pc6,'Personen met WW, Bijstand en/of AO uitkering\nBeneden AOW-leeftijd']
        df.at[i[0],'Huurwoning'] = df_cbs.at[pc6,'Huurwoning']
df
```

```python
# Schaalverdeling bouwjaar

df = df.copy()

for i in df.iterrows():
    bouwjaar = df.loc[i[0],'bouwjaar']
    if bouwjaar <= 1940:
        df.at[i[0],'Bouwjaar_schaal'] = 5
    if bouwjaar > 1940 and bouwjaar <= 1960:
        df.at[i[0],'Bouwjaar_schaal'] = 4
    if bouwjaar > 1960 and bouwjaar <= 1980:
        df.at[i[0],'Bouwjaar_schaal'] = 3
    if bouwjaar > 1980 and bouwjaar <= 2000:
        df.at[i[0],'Bouwjaar_schaal'] = 2
    if bouwjaar > 2000:
        df.at[i[0],'Bouwjaar_schaal'] = 1
```
