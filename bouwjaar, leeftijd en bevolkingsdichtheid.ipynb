```python
import pandas as pd 
import numpy as np 
import matplotlib.pyplot as plt 
```

```python
# Get list of all neighbor names of Amsterdam.
df_code = pd.read_csv('../Data/cbs/brt2020.csv', delimiter = ';')
```

```python
df_code = df_code[df_code.GM_NAAM == 'Amsterdam']
df_code
```

```python
buurten = list(df_code['buurtnaam2020'])
wijken = df_code['WK_NAAM'].unique()
print(len(buurten))
print(len(wijken))
```

```python
# Get age
df_CBS = pd.read_csv('../Data/CBS_1.csv', delimiter = ';')
df_CBS
```

```python
df_buurt = df_CBS[df_CBS.SoortRegio_2 == 'Buurt     ']
df_buurt = df_buurt[['Codering_3', 'AantalInwoners_5', 'k_0Tot15Jaar_8', 'k_65JaarOfOuder_12', 'GemiddeldeHuishoudensgrootte_32', 'Bevolkingsdichtheid_33']]
df_buurt.rename(columns={'Codering_3': 'BuurtCode', 'AantalInwoners_5': 'InwonersTotaal', 'k_0Tot15Jaar_8': 'Leeftijd 0-15', 'k_65JaarOfOuder_12': 'Leeftijd 65+', 'GemiddeldeHuishoudensgrootte_32' : 'GemiddeldeHuishoudensgrootte', 'Bevolkingsdichtheid_33' : 'Bevolkingsdichtheid'}, inplace=True)
df_buurt
```

```python
df_buurt.insert(3, '% Leeftijd 0-15', (df_buurt['Leeftijd 0-15'] / df_buurt['InwonersTotaal']) *100)
df_buurt.insert(5, '% Leeftijd 65+', (df_buurt['Leeftijd 65+'] / df_buurt['InwonersTotaal']) *100)
df_buurt.insert(1, 'BuurtNaam', buurten)
```

```python
df_buurt
```

```python
df_buurt.to_csv('buurt_leeftijd_bevolkingsdichtheid.csv')
```

```python
# Get building year
df_bouwjaar = pd.read_csv('../Data/Kerncijfers_wijken_en_buurten_2020.csv', delimiter = ';', skiprows = 3)
df_bouwjaar = df_bouwjaar[['Wijken en buurten', 'Wonen|Woningen naar bouwjaar|Bouwjaar voor 2000' , 'Wonen|Woningen naar bouwjaar|Bouwjaar vanaf 2000']]
df_bouwjaar.rename(columns={'Wonen|Woningen naar bouwjaar|Bouwjaar voor 2000' : 'Bouwjaar voor 2000', 'Wonen|Woningen naar bouwjaar|Bouwjaar vanaf 2000' : 'Bouwjaar vanaf 2000'}, inplace=True)
df_bouwjaar
```

```python
df_bouwjaar_buurt = df_bouwjaar[~df_bouwjaar['Wijken en buurten'].isin(wijken)]
df_bouwjaar_buurt


```

```python
df_bouwjaar_buurt.to_csv('buurt_bouwjaar.csv')
```
